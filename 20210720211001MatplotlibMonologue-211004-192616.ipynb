```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,6,10)
```

```python
x
```

```python
y = x**2
```

```python
y
```

```python
#plot the y vs x - parabola
```

```python
#function based method
plt.plot(x,y)
plt.xlabel('this is the x value')
plt.ylabel('this is the y value')
plt.title('Chart Title')
```

```python
plt.subplot(2,1,1) # 1 row and 2 columns and the plot number
plt.plot(x,y,'r')
plt.xlabel('x value')
plt.ylabel('y value')
plt.title('Chart 1')

plt.subplot(2,1,2) # 1 row and 2 columns and the plot number
plt.plot(y,x,'b')
plt.xlabel('y value')
plt.ylabel('x value')
plt.title('Chart 2')
```

```python
plt.subplot(2,2,1) # 1 row and 2 columns and the plot number
plt.plot(x,y,'r')
plt.xlabel('x value')
plt.ylabel('y value')
plt.title('Chart 1')

plt.subplot(2,2,2) # 1 row and 2 columns and the plot number
plt.plot(y,x,'b')
plt.xlabel('y value')
plt.ylabel('x value')
plt.title('Chart 2')

plt.subplot(2,2,3) # 1 row and 2 columns and the plot number
plt.plot(x,x**3,'r')
plt.xlabel('x value')
plt.ylabel('cube')
plt.title('Chart 3')

plt.subplot(2,2,4) # 1 row and 2 columns and the plot number
plt.plot(x**3,x,'b')
plt.xlabel('cube')
plt.ylabel('x value')
plt.title('Chart 4')
```

```python
#object based method
#

#get a canvas
fig = plt.figure() #blank canvas
axis = fig.add_axes([0.1,0.1,0.8,0.8]) #%left, %bottom, %length, %height
axis.plot(x,y)
axis.set_xlabel('X value')
axis.set_ylabel('Y value')
axis.set_title('Chart Title')
```

```python
fig = plt.figure() #blank canvas
axis1 = fig.add_axes([0.1,0.1,0.8,0.8]) #%left, %bottom, %length, %height
axis1.plot(x,y)
axis1.set_xlabel('X value')
axis1.set_ylabel('Y value')
axis1.set_title('Big Title')

axis2 = fig.add_axes([0.2,0.5,0.3,0.3]) #%left, %bottom, %length, %height
axis2.plot(y,x)
axis2.set_xlabel('Y value')
axis2.set_ylabel('X value')
axis2.set_title('Small Chart')
```

```python
fig = plt.figure() #blank canvas
axis = fig.add_axes([0.1,0.1,0.8,0.8]) #%left, %bottom, %length, %height
axis.plot(x,x**2, label = 'sqr')
axis.plot(x,x**3, label = 'cube')
axis.legend()
```
