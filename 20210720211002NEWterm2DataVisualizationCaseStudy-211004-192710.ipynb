### **5.1 Bar Chart**

- Bar chart is one of the simplest tool for data visualization because they **map categories** to **numbers**.


- They are useful for **categorical data** that doesn’t have a lot of different categories.

- For pandas's barplot official documentation click <a href="https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.bar.html">**here**</a>.

```python
figure = plt.figure(figsize=[15, 7]) #blank canvas, size

#df_mcd['Category'].value_counts().plot(kind = 'bar') #plot command, type of plot
df_mcd['Category'].value_counts().plot.bar()

plt.xticks(rotation=90, size=12)
plt.yticks(ticks=np.arange(0, 20, 2), size=12)

plt.xlabel(xlabel='Category', size=14)
plt.ylabel(ylabel='Number of Items', size=14)
plt.title(label='Number of Items per Category', size=16)
plt.show()

#show the numbers on the bar chart
```

```python
figure = plt.figure(figsize=[15, 7])

df_mcd['High_Or_Low_Sugar'].value_counts().plot(kind = 'bar', color='#1B9E77')


plt.xticks(rotation=90, size=12)
plt.yticks(ticks=np.arange(0,51, 5), size=12)
plt.xlabel(xlabel='High_Or_Low_Sugar', size=14)
plt.ylabel(ylabel='Number of Items', size=14)
plt.title(label='High_Or_Low_Sugar', size=16)
plt.show()
```

**Observation:**

- We can observe that the **Cold Beverages** category is the **highest selling category** amongst other categories.

- What if we want to know the **percentage** of total **Cold Beverages** or **Desserts**?


- This bar chart plots the __absolute numbers__, but we can also plot the __relative proportions__ of each category.

```python
# Extracting labels of Category in a list
x_value = df_mcd['Category'].value_counts().index.to_list()

# Extracting proportion values for each category
y_value = df_mcd['Category'].value_counts().values / len(df_mcd)

# Display the extracted values
print('x-axis values:', x_value)
print('y-axis values', y_value)
```

```python
# Creating a custom figure size of 15 X 7 inches
figure = plt.figure(figsize=[15, 7])

# Using magic of pandas bar function
(df_mcd['Category'].value_counts() / len(df_mcd)).plot.bar(color='#3F617A')

# Changing x-ticks label size to 12 and rotating to 90 degrees
plt.xticks(rotation=90, size=12)

# Changing y-ticks value using an array of size 0.3 with a step size of 0.05
plt.yticks(ticks=np.arange(0, 0.3, 0.05), size=12)

# Labelling x-axis with a custom label and size of 14
plt.xlabel(xlabel='Category', size=14)

# Labelling y-axis with a custom label and size of 14
plt.ylabel(ylabel='Number of Items', size=14)

# Labelling title with a custom label and size of 16
plt.title(label='Number of Items per Category', size=16)

# Dispaly the output by rendering visual on the screen
plt.show()
```

**Observation:**

- We can observe that **60% items** belong to the **Cold Beverages, Desserts** and **Sandwiches and Wraps**.

### **5.4 Histogram**

- It is a graphical **display** of **data** using **bars** of different heights where each bar **groups numbers into ranges**. 

- **Taller bars** show that **more data** falls in that range. They are **useful** in **identifying** the **distribution** of features.

- For pandas hist plot official documentation click <a href="https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.hist.html">**here**</a>.

```python

figure = plt.figure(figsize=[15, 7])

my_choice_of_bins = list(np.arange(0,501,50))

df_mcd['Serve_Size'].plot.hist(bins=my_choice_of_bins, color='brown')

plt.xticks(ticks=my_choice_of_bins, size=12)
plt.yticks(ticks=np.arange(0, 22, 3), size=12)
plt.xlabel(xlabel='Serve Size', size=14)
plt.ylabel(ylabel='Frequency', size=14)
plt.title(label='Distribution of Serve Size Feature', size=16)
plt.show()
```

```python
# Creating a custom figure size of 15 X 7 inches
figure = plt.figure(figsize=[15, 7])

# Creating custom bins for the graph
bins = list(np.arange(0,501,50))

# Using magic of pandas area() function
df_mcd['Serve_Size'].plot.hist(bins=bins, color='brown')
# Changing x-ticks equal to custom bin values
plt.xticks(ticks=bins, size=12)

# Changing y-ticks ranging from 0 to 22 with step size of 3 and label size of 12
plt.yticks(ticks=np.arange(0, 22, 3), size=12)

# Labelling x-axis with a custom label and size of 14
plt.xlabel(xlabel='Serve Size', size=14)

# Labelling y-axis with a custom label and size of 14
plt.ylabel(ylabel='Frequency', size=14)

# Labelling title with a custom label and size of 14
plt.title(label='Distribution of Serve Size Feature', size=16)

# Dispaly the output by rendering visual on the screen
plt.show()
```

### **5.5 KDE Plot**

- This graph helps in visualizing the **distribution of data** over a **continuous interval** or **time period**. 


- This chart is a **variation of a Histogram** that uses **smoothing** technique to plot values. 


- The **peaks** of a **density plot** help display where **values are concentrated** over the **interval**.


- **Advantage:** Better at determining the distribution shape because they're **unaffected** by **bins size**.

- For pandas kde plot official documentation click <a href="https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.kde.html">**here**</a>.

```python
# Creating a custom figure size of 15 X 7 inches
figure = plt.figure(figsize=[15, 7])

# Using magic of pandas kde() function
df_mcd['Serve_Size'].plot.kde(color='black')

# Changing x-ticks ranging from -200 to 800 with step size of 50 and label size of 12
plt.xticks(ticks=np.arange(-200, 800, 50), size=12)

# Changing y-ticks labels size to 12
plt.yticks(size=12)

# Labelling x-axis with a custom label and size of 14
plt.xlabel(xlabel='Serve Size', size=14)

# Labelling y-axis with a custom label and size of 14
plt.ylabel(ylabel='Density', size=14)

# Labelling title with a custom label and size of 14
plt.title(label='Density of Serve Size Feature', size=16)

# Dispaly the output by rendering visual on the screen
plt.show()
```

**Observation:**

- The data in **Serve Size** column is somewhat following a **normal distribution**.

### **5.6 Donut Chart**

- This graph is a **variation of a pie chart** except that it is hollow in the center which makes it look like a donut.

- The pie charts are **sometimes criticised** for focusing readers on the **proportional areas** of the slices to one another.

- This makes it tricky to see the differences between slices, especially when you try to __compare multiple pie charts together__.

- A donut chart somewhat remedies this problem by **de-emphasizing** the use of the area. 

- Instead, readers focus more on reading the __length of the arcs__, rather than __comparing__ the proportions between __slices__.

### **7.4 Parallel Coordinates**

- This type of visualisation is used for plotting multivariate, numerical data.

- They are **ideal** for **comparing many variables together** and understanding the relationships between them.

- Each variable is given its own axis and all the axes are placed in parallel to each other.

- They can **have** a **different or common scale** for each variable to display the unit of measurement.

- Values are plotted as a **series of lines** that connected across all the axes.

- This means that **each line** is a **collection of points** placed on each axis, that have all been connected together.

- For pandas's parallel_coordinates official documentation click <a href="https://pandas.pydata.org/docs/reference/api/pandas.plotting.parallel_coordinates.html">**here**</a>.

- Before plotting the data we can normalize all the variables using a scaling method.

- We will be using min-max scaling: $\large\frac{x-min(x_i)}{max(x_i)-min(x_i)}$

```python
# Copy original data to an object
mcd_normal = df_mcd.copy()

# Initiate a list of categorical features
cat_feats = ['Category', 'Item', 'Sodium', 'High_Or_Low_Sugar']

# Iterate over the data variables to perform scaling
for col in mcd_normal.columns:

  # Check if the categorical variables are present in the dataset
  if col not in cat_feats:
    
    # Perform min-max scaling
    mcd_normal[col] = (mcd_normal[col] - mcd_normal[col].min()) / (mcd_normal[col].max() - mcd_normal[col].min())

# Dispaly top 5 rows
mcd_normal.head()
```

```python
# Creating a custom figure size of 15 X 7 inches
figure = plt.figure(figsize=[15, 7])

# Using magic of pandas's parallel_coordinates() function
pd.plotting.parallel_coordinates(frame=mcd_normal, class_column='High_Or_Low_Sugar', 
                                  cols=['Energy', 'Protein', 'Total_Fat', 'Carbohydrates'])

# Changing x-ticks label size to 12
plt.xticks(size=12)

# Changing y-ticks label size to 12
plt.yticks(size=12)

# Labelling x-axis with a custom label and size of 14
plt.xlabel(xlabel='Numerical Features', size=14)

# Labelling y-axis with a custom label and size of 14
plt.ylabel(ylabel='Values', size=14)

# Labelling title with a custom label and size of 14
plt.title(label='Energy, Protein, Total Fat, Carbohydrates vs Sugar Levels', size=16)

# Dispaly the output by rendering visual on the screen
plt.show()
```

### **7.5 Violin Plot**

- This plot is used to **visualise** the **distribution** of the data and its probability density.

- This plot is a **combination** of a **Box Plot** and a **Density Plot** that is rotated and placed on each side.

- The **white dot** in the **middle** is the **median** value and the **thick black bar** in the center represents the **interquartile range**.

- The thin black line extended from it represents the upper (max) and lower (min) adjacent values in the data. 

- **Sometimes** the graph **marker** is **clipped** from the **end of the line**.

- Box plots tends to limit their data display because they hide the distribution of the data.

- For example, with Box Plots, you can't see if the distribution is bimodal or multimodal. 

- For seaborn's violinplot official documentation click <a href="https://seaborn.pydata.org/generated/seaborn.violinplot.html">**here**</a>.

```python
# Creating a custom figure size of 15 X 7 inches
figure = plt.figure(figsize=[15, 7])

# Using magic of seaborn's violinplot() function
sns.violinplot(x='High_Or_Low_Sugar', y='Energy', data=df_mcd, palette='magma')

# Changing x-ticks label size to 12
plt.xticks(size=12)

# Changing y-ticks ranging from -200 to 1001 with step size of 50 and label size of 12
plt.yticks(ticks=np.arange(-200, 1001, 50), size=12)

# Labelling x-axis with a custom label and size of 14
plt.xlabel(xlabel='Sugar Level', size=14)

# Labelling y-axis with a custom label and size of 14
plt.ylabel(ylabel='Energy', size=14)

# Labelling title with a custom label and size of 14
plt.title(label='Energy Variation with Sugar', size=16)

# Enable grid
plt.grid(b=True, axis='y')

# Dispaly the output by rendering visual on the screen
plt.show()
```

**Observation:**

- From the above violin plot we can clearly see the distribution on the basis of Sugar Level and Energy.

- The **items** having **high sugar** content have **lower energy** when compared to items having low sugar content.
